```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt  

# Load your data from the Excel sheet
data = pd.read_csv(r"F:\Miami\RefinedVC\class400.csv")

# Get unique classes in the 'class' column
unique_classes = data['Class'].unique()

# Initialize an empty list to store RMSE values for each class
rmse_values_by_class = []

# Iterate over each class and calculate RMSE
for class_label in unique_classes:
    class_data = data[data['Class'] == class_label]
    
    # Assuming 'NAVD88(GCPs)' is your actual data column, and 'Lidar' is your predicted data column
    y_actual = class_data['NAVD88']
    y_predicted = class_data['RASTERVALU']

    # Calculate the Mean Squared Error (MSE)
    MSE = np.square(np.subtract(y_actual, y_predicted)).mean()

    # Calculate the Root Mean Square Error (RMSE)
    RMSE = math.sqrt(MSE)
    
    # Append RMSE value to the list
    rmse_values_by_class.append(RMSE)

    print(f"RMSE for Class {class_label}: {RMSE}")

# Create a histogram of RMSE values by class
plt.hist(rmse_values_by_class, bins=30, color='blue', edgecolor='black')
plt.xlabel('RMSE Value')
plt.ylabel('Frequency')
plt.title('Histogram of RMSE by Class')
plt.grid(False)

# Show the histogram
plt.show()
```

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns  # Import seaborn for boxplots

# Load your data from the Excel sheet
data = pd.read_excel(r"F:\Thesis_Data\data_vc\test_try.xlsx")

# Get unique classes in the 'Class' column
unique_classes = data['Class'].unique()

# Initialize an empty list to store RMSE values for each class
rmse_values_by_class = []

# Iterate over each class and calculate RMSE
for class_label in unique_classes:
    class_data = data[data['Class'] == class_label]
    
    # Assuming 'NAVD88(GCPs)' is your actual data column, and 'Lidar' is your predicted data column
    y_actual = class_data['NAVD88']
    y_predicted = class_data['Lidar']

    # Calculate the Mean Squared Error (MSE)
    MSE = np.square(np.subtract(y_actual, y_predicted)).mean()

    # Calculate the Root Mean Square Error (RMSE)
    RMSE = math.sqrt(MSE)
    
    # Append RMSE value to the list
    rmse_values_by_class.append({'Class': class_label, 'RMSE': RMSE})

    print(f"RMSE for Class {class_label}: {RMSE}")

# Convert the list of dictionaries to a DataFrame
rmse_df = pd.DataFrame(rmse_values_by_class)

# Create a box-and-whisker plot with lower and upper values
plt.figure(figsize=(12, 8))
sns.boxplot(x='Class', y='RMSE', data=rmse_df, showfliers=False)  # Hide outliers for better clarity

# Calculate and plot lower and upper RMSE values for each class
for i, class_label in enumerate(unique_classes):
    class_data = rmse_df[rmse_df['Class'] == class_label]
    lower_rmse = class_data['RMSE'].quantile(0.25)  # Lower quartile (25th percentile)
    upper_rmse = class_data['RMSE'].quantile(0.75)  # Upper quartile (75th percentile)
    plt.scatter(i, lower_rmse, color='red', marker='_', s=200)  # Plot lower RMSE value
    plt.scatter(i, upper_rmse, color='blue', marker='_', s=200)   # Plot upper RMSE value

plt.xlabel('Class')
plt.ylabel('RMSE Value')
plt.title('Box-and-Whisker Plot of RMSE by Class with Lower and Upper Values')
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
import math

# Load your data from the Excel sheet
data = pd.read_excel("F:\Thesis_Data\RMSE.xlsx")

# Assuming 'NAVD88(GCPs)' is your target variable, and 'DEM' is your predictor variable
X = data[['noaa']]
y = data['usgs']

# Create a linear regression model
model = LinearRegression()

# Fit the model to your data
model.fit(X, y)

# Predict 'NAVD88(GCPs)' based on 'DEM' values
y_predicted = model.predict(X)

# Calculate the Mean Squared Error (MSE)
MSE = np.square(np.subtract(y, y_predicted)).mean()

# Calculate the Root Mean Square Error (RMSE)
RMSE = math.sqrt(MSE)

# Calculate the R-squared value
r_squared = model.score(X, y)

# Plot the original data points and the regression line
plt.scatter(X, y, label='Actual Data', color='blue')
plt.plot(X, y_predicted, label='Linear Regression', color='red')
plt.xlabel('noaa')
plt.ylabel('usgs')
plt.title('Linear Regression\nRMSE = {:.2f}, R-squared = {:.2f}'.format(RMSE, r_squared))
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

# Get the coefficients of the linear regression equation
intercept = model.intercept_
slope = model.coef_[0]
print(f'Intercept: {intercept}')
print(f'Slope: {slope}')
print(f'RMSE: {RMSE}')
print(f'R-squared: {r_squared}')


```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
import math

# Load your data from the Excel sheet
data = pd.read_excel("F:\Thesis_Data\RMSE1.xlsx")

# Get unique land cover types (assuming 'id' represents the land cover types)
land_cover_types = data['id'].unique()

# Initialize a dictionary to store the results for each land cover type
results = {}

# Loop through each land cover type and fit a linear regression model
for land_cover_type in land_cover_types:
    # Filter the data for the current land cover type
    subset_data = data[data['id'] == land_cover_type]
    
    # Assuming 'noaa' is your predictor variable and 'usgs' is the target variable
    X = subset_data[['noaa']]
    y = subset_data['usgs']
    
    # Create a linear regression model
    model = LinearRegression()
    
    # Fit the model to your data
    model.fit(X, y)
    
    # Predict 'usgs' based on 'noaa' values
    y_predicted = model.predict(X)
    
    # Calculate the Mean Squared Error (MSE)
    MSE = np.square(np.subtract(y, y_predicted)).mean()
    
    # Calculate the Root Mean Square Error (RMSE)
    RMSE = math.sqrt(MSE)
    
    # Calculate the R-squared value
    r_squared = model.score(X, y)
    
    # Store the results in the dictionary
    results[land_cover_type] = {
        'Intercept': model.intercept_,
        'Slope': model.coef_[0],
        'RMSE': RMSE,
        'R-squared': r_squared
    }
    
    # Plot the original data points and the regression line for the current land cover type
    plt.scatter(X, y, label=f'Actual Data ({land_cover_type})', alpha=0.7)
    plt.plot(X, y_predicted, label=f'Linear Regression ({land_cover_type})')
    
# Customize the plot
plt.xlabel('noaa')
plt.ylabel('usgs')
plt.title('Linear Regression for Different Land Cover Types')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

# Print the results for each land cover type
for land_cover_type, result in results.items():
    print(f'Land Cover Type: {land_cover_type}')
    print(f'Intercept: {result["Intercept"]}')
    print(f'Slope: {result["Slope"]}')
    print(f'RMSE: {result["RMSE"]}')
    print(f'R-squared: {result["R-squared"]}')
    print('-' * 30)
```
